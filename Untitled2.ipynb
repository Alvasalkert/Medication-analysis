```python
#creating measurement_mg collumn to later on be able to do math later
import re
import pandas as pd
import numpy as np

df1 = pd.DataFrame({'name': df_all.name, 
                   'pprice': df_all.pprice,
                   'sprice': df_all.sprice,
                   'form_eng': df_all.form_eng})


df1['measurement_MG'] = df1['name'].apply(lambda x: re.search(r'([\d]+)(?:MCG|MG)', x).group() 
if re.search(r'([\d]+)(?:MCG|MG)', x) else np.nan)

df1 = df1.dropna()

display(df1)
```

```python
#separating the numbers from the units measurement in the measurment mg collumn 
df1['meas_num_mcg'] = df1['measurement_MG'].apply(lambda x: float(x.split("MG")[0]) if x.endswith("MG") else float(x.split("MCG")[0]) / 100)
df1['sprice'] = df1['sprice'].astype('float')
df1['pprice'] = df1['pprice'].astype('float')
df1['meas_num_mcg'] = df1['meas_num_mcg'].astype('float')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
    print(df1)
```

```python
#finalizing with getting the markup and the price per mg for every medication 
df1['meas_num_mcg'].apply(lambda x: float(x))
df1['pprice'].apply(lambda x: float(x))
df1["sprice"].apply(lambda x: float(x))

df1["price/mg"] = df1.sprice / df1.meas_num_mcg
df1["mark_up%"] = (df1.sprice - df1.pprice) / df1.pprice * 100
display(df1)
```
